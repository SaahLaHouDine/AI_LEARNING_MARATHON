# SECTION2:BASIC IMPLAMENTATION


```
import numpy as np
V1=np.array([1 , -5]) # array are typically list au number here vectors i just create 5 vectors and add them together
V2=np.array([7,8])
V3=np.array([4,3])
V4=np.array([2,2])
V5=np.array([2,2])
V_SUM= V1 + V2+ V3 + V4+V5
print(f"V1+V2+V3+V4+V5={V_SUM}")
```

# SECTION 3: VISUALISATION

```
import matplotlib.pyplot  as ptl
#drawing a vectors 2 vectoors in a orthonormal coordinate system
ptl.figure(figsize=(6,6))
ptl.quiver (0,0,V1[0], V1[1],angles='xy',scale_units='xy',scale=1,color='red')
ptl.quiver(0,0,V2[0], V2[1],angles='xy',scale_units='xy',scale=1,color='blue')
ptl.xlim(-10,10)
ptl.ylim(-10,10)
ptl.grid()
ptl.show()
```
